```python
#                  *****************************Question 1*******************************
# ----------------------------------------------------------------------------------------------------------------


#1.	Suppose the cover price of a book is Rs. 200, but bookstores get a 25% discount. Shipping costs Rs. 40 for the first copy 
# and Rs. 10 for each additional copy. Write a Python program to calculate the total wholesale cost for 60 copies

TotalNumberOfCopies = 60
PricePerCopies = 200
ShippingFirstCopy = 40
ShippingRestCopy = 10
BookStoreDiscount  = 25/100

PriceAfterDiscount = PricePerCopies - (PricePerCopies * BookStoreDiscount)
PriceOfTheTotalCopies = PriceAfterDiscount * TotalNumberOfCopies
PriceAfterShipping = ShippingFirstCopy + ((TotalNumberOfCopies - 1) * ShippingRestCopy)

print ('The total price for 60 copies with the discount and shipping charges are ')
print ('Total price of the books is: ' + str(PriceOfTheTotalCopies))
print ('Total Shipping charges: ' + str(PriceAfterShipping))
print ('Total price is: ' + str(PriceOfTheTotalCopies + PriceAfterShipping))

```

```python
#                  *****************************Question 2*******************************
# ----------------------------------------------------------------------------------------------------------------


#2.A number, x, is a power of y if it is divisible by y and x/y is a power of y. 
# Write a function in Python that takes parameters x and y and returns True if x is a power of y.

# Method to check whether the provided number i.e. x is the power or y
def checkPower(baseValue, powerValue): 

    if baseValue == powerValue:
        return True
    
    storeTempValue = baseValue

    while (storeTempValue <= powerValue):
        if storeTempValue == powerValue:
            return True
        storeTempValue *= baseValue
    return False

powerValue = str(input("Please enter a positive number for X  ->  "))
baseValue = str(input("Please enter a positive number of Y  ->  "))

if powerValue.isdigit() == 0 or baseValue.isdigit() == 0 :
    print ("\n\n********You have typed the non negative number. Please enter the positive number**********")

checkPower(int(baseValue), int(powerValue))

```

```python
#                  *****************************Question 5*******************************
# ----------------------------------------------------------------------------------------------------------------

# 5.Write a function in Python that takes three integers as arguments, and that prints either “Yes” or “No”, 
# depending on whether you can or cannot form a triangle from the given lengths

def checkLength(first, second, third):
    #Trying to add two length numnbers and storing it in the variable
    L1 = first + second
    L2 = first + third
    L3 = second + third

    if first < L3 and second < L2 and third < L1 :
        print ("Yes we can draw the triangle")
    else:
        print ("No we cannot draw the triangle")
    

firstLength = str(input("\nPlease enter a positive number for first length  -> "))
secondLength = str(input("Please enter a positive number for second length  -> "))
thirdLength = str(input("Please enter a positive number for third length  -> "))

if firstLength.isdigit() == 0 or secondLength.isdigit() == 0 or thirdLength.isdigit() ==0:
    print ("\n\n********You have typed the non negative number. Please enter the positive number**********")

checkLength(int(firstLength), int(secondLength), int(thirdLength))

```

```python
#                  *****************************Question 7*******************************
# ----------------------------------------------------------------------------------------------------------------

# 7.Write a function in Python that takes a list of numbers and returns the cumulative sum; that is, a new list where 
# the nth element is the sum of the first n + 1 elements from the original list.
# list = [ 1, 2, 3, 4 ]

def cumulativeSum(values):
    finalValue = []
    sumOfValue = 0
    for i in values:
        sumOfValue = sumOfValue + i
        finalValue.append(sumOfValue)
    print (finalValue)

values = [1, 2, 3, 4, 5, 6]
    
cumulativeSum(values)
```

```python
#                  *****************************Question 8*******************************
# ----------------------------------------------------------------------------------------------------------------

# 8.Write a Python program that reads a file, breaks each line into words, strips whitespace and punctuation from the words, 
# and converts them to uppercase.
# Please keep the file i.e. SomeText in this path C:/Downloads

# File contain this string "News is information about current events. This may %be provided through many #different media: word of mouth, printing,"

import string
fileContent = open('C:/Downloads/SomeText', 'r')

words = fileContent.read()
print(words)
splitWord = words.split(' ')
print (splitWord)
finalWord = ''
for eachWord in splitWord:
    finalWord = finalWord + ' '
    for eachChar in eachWord:
        if (eachChar in string.punctuation) or (eachChar in string.whitespace):
            pass            
        else:
            finalWord = finalWord + eachChar.upper()

print ("\n",finalWord)

fileContent.close()
```

```python
#                  *****************************Question 9*******************************
# ----------------------------------------------------------------------------------------------------------------


# 9.Write a program that searches a directory and all of its subdirectories recursively, and returns a list of complete paths 
# for all files with a given suffix (like .mp3)
# Please keep the attached mp3 files at C:/Downlads path location, so that it'll fetch accordingly

import os
# print (os.getcwd())
# print(os.walk('D:/Files/test/'))
def readFilesFromDirectory(dirname):
    files = []
    finalFile = []
    for root, dirs, files in os.walk(dirname):        
        for filename in files:
#             print(filename,'\n', filename[-4:])
            if filename[-4] == ".mp3":        
                finalFile.append(os.path.join(root, filename))
    return finalFile

readFilesFromDirectory('C:/Downloads/')
```
